## Exploring the Wine Dataset: Dimensionality Reduction with PCA for Enhanced Visualization and Analysis

This project demonstrates the application of Principal Component Analysis (PCA) to the Wine dataset, a classic dataset in machine learning and statistics. The aim is to explore and visualize high-dimensional data by reducing its dimensionality, making it easier to understand and interpret. PCA is used to transform the data into a lower-dimensional space while preserving as much variance as possible, allowing us to visualize complex data structures in 2D and 3D.

## Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
```

## Load the Wine dataset

**What is the Wine Dataset?**

The Wine dataset is a classic dataset often used in machine learning and statistics for classification and regression tasks. It contains chemical analysis data of wines grown in the same region in Italy but derived from three different cultivars (types of grapes).

**Features:**

* **Alcohol**: Alcohol content of the wine.
* **Malic Acid**: Amount of malic acid in the wine.
* **Ash**: Ash content in the wine.
* **Alcalinity of Ash**: Measure of the alkalinity of ash in the wine.
* **Magnesium**: Magnesium content in the wine.
* **Total Phenols**: Total phenolic content in the wine.
* **Flavanoids**: Flavanoid content in the wine.
* **Nonflavanoid Phenols**: Non-flavanoid phenolic content in the wine.
* **Proanthocyanins**: Proanthocyanin content in the wine.
* **Color Intensity**: Intensity of color in the wine.
* **Hue**: Hue of the wine color.
* **OD280/OD315 of Diluted Wines**: Proportion of diluted wines.
* **Proline**: Proline content in the wine.

```python
wine = datasets.load_wine()
X = wine.data
y = wine.target
```

```python
print("data points :\n", X)
print('\n')
print("target points :\n", y)
```

```python
# Get the shape of the data
num_samples, num_features = wine.data.shape

# Print the number of features and number of samples
print("Number of features:", num_features)
print("number of samples :", num_samples)
```

## Standardize the data

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```

# What is Principal Component Analysis (PCA)

Principal Component Analysis (PCA) is a widely-used dimensionality reduction technique in statistics and machine learning. It transforms high-dimensional data into a lower-dimensional form while preserving as much variance (information) as possible. PCA is particularly useful for visualizing and analyzing high-dimensional data.

## Apply PCA for 2D visualization

Applying PCA for 2D visualization involves reducing the dimensionality of a high-dimensional dataset to two dimensions using Principal Component Analysis (PCA). This allows us to visualize complex data in a 2-dimensional space, making it easier to interpret and analyze.
                                                                                              
**Why Apply PCA for 2D Visualization?**

1. `Simplification of High-Dimensional Data`:

   High-dimensional data can be difficult to visualize and interpret because we cannot directly visualize more than three dimensions. PCA helps simplify this by projecting the data onto a 2-dimensional plane, retaining as much of the data's variance (information) as possible.

2. `Data Understanding`:

   Visualizing data in 2D can help identify patterns, clusters, outliers, and relationships between data points that may not be apparent in high-dimensional space.

3. `Feature Reduction`:

   Reducing the number of features can help in identifying the most important features and understanding their impact on the dataset's variability.

```python
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X_scaled)
print(X_pca_2d[:10])
```

## Apply PCA for 3D visualization

Applying PCA for 3D visualization involves reducing the dimensionality of a high-dimensional dataset to three dimensions using Principal Component Analysis (PCA). This allows us to visualize complex data in a 3-dimensional space, making it easier to interpret and analyze.

**Why Apply PCA for 2D Visualization?**

1. `Simplification of High-Dimensional Data`:

    High-dimensional data is challenging to visualize and interpret because we cannot directly observe more than three dimensions. PCA simplifies this by projecting the data onto a 2-dimensional plane, retaining as much of the original variance (information) as possible. This makes complex datasets easier to handle and understand.

2. `Data Understanding`:

    Visualizing data in 2D can reveal patterns, clusters, and relationships between data points that may not be visible in higher dimensions. By reducing the data to two principal components, you can identify structural features, such as groupings and separations, more clearly.

3. `Feature Reduction`:

    Reducing the number of features helps focus on the most important aspects of the data. PCA identifies key features that contribute most to the variance in the dataset, making it easier to understand their impact and simplify subsequent analysis or modeling.

```python
pca_3d = PCA(n_components=3)
X_pca_3d = pca_3d.fit_transform(X_scaled)
print(X_pca_3d[:10])
```

## 2D Plot data samples with lables

```python
plt.figure(figsize=(8, 6))
for i, target_name in enumerate(wine.target_names):
    plt.scatter(X_pca_2d[y == i, 0], X_pca_2d[y == i, 1], label=target_name)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.title('2D PCA on Wine Dataset')
plt.show()
```

## 3D Plot data samples with lables

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
for i, target_name in enumerate(wine.target_names):
    ax.scatter(X_pca_3d[y == i, 0], X_pca_3d[y == i, 1], X_pca_3d[y == i, 2], label=target_name)
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
plt.title('3D PCA on Wine Dataset')
plt.legend()
plt.show()
```

## 3D Plot with diffrent angles

```python
# Create a figure with 4 subplots for different angles
fig, axs = plt.subplots(2, 2, subplot_kw={'projection': '3d'}, figsize=(12, 10))

# Define viewing angles for the subplots
angles = [(30, 30), (60, 45), (90, 60), (120, 90)]

# Plot each subplot with a different angle
for ax, angle in zip(axs.flatten(), angles):
    for i, target_name in enumerate(wine.target_names):
        ax.scatter(X_pca_3d[y == i, 0], X_pca_3d[y == i, 1], X_pca_3d[y == i, 2], label=target_name)
    ax.set_xlabel('Principal Component 1')
    ax.set_ylabel('Principal Component 2')
    ax.set_zlabel('Principal Component 3')
    ax.view_init(elev=angle[0], azim=angle[1])
    ax.set_title(f'View Angle: Elevation {angle[0]}, Azimuth {angle[1]}')
    ax.legend(loc='best')

plt.tight_layout()
plt.show()
```
